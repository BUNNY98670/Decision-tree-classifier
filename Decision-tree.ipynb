DECISION TREE CLASSFIER

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier   # for classification
from sklearn.tree import DecisionTreeRegressor    # for regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import plot_tree
```

```python

from sklearn.datasets import load_iris

iris = load_iris()

print("iris['DESCR']:", iris['DESCR'])
```

```python
print(iris['target'])
```

```python
%matplotlib inline
```

```python
x = pd.DataFrame(
    iris['data'],
    columns=[
        'sepal length (cm)',
        'sepal width (cm)',
        'petal length (cm)',
        'petal width (cm)'
    ]
)
```

```python
print(x.head())
```

```python
y= iris['target']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
##applying the decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,6))

tree.plot_tree(treeclassifier,filled=True,feature_names=x.columns,
               class_names=iris['target_names'],max_depth=3
               
)
plt.show()
```

```python
y_pred = treeclassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
cm= confusion_matrix(y_test, y_pred)
print(cm)

print(classification_report(y_test, y_pred))
```

```python
prams = {
    'criterion': ['gini', 'entropy','log_loss'],
    'max_depth': [1,23, 4, 5, 6],
    'splitter':  ['best', 'random']  , 
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
prams
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(estimator=treemodel, param_grid=prams, cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score (y_test,y_pred)
```
